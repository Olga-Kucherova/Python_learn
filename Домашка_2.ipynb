Для следующей задачи нам нужно понять, как проверить что какой-то элемент есть в списке.
Для этого мы используем конструкцию in
```
if элемент in список:
  делаем действие
else:
  делаем другое действие
```



Посмотрите на пример ниже

````python
buys = ["apple", "pineapple", "blueberry"]
if "apple" in buys:
  print("We can buy apple!!!")
else:
  print("We can't buy apple")
if "ice-cream" in buys:
  print("We can buy ice-cream!!!")
else:
  print("We can't buy ice-cream")
````

9. У нас есть список гостей. С клавиатуры вводится имя пришедшего, если он есть в списке, мы его пускаем, если нет, то мы его не пускаем!

На паре мы не успели познакомиться с логическим отрицанием.
"Not" - это логический оператор в языке программирования Python. Он работает с булевыми значениями, которые могут быть истинными или ложными.

"Not" (англ. "не") возвращает противоположное значение в зависимости от исходного значения. Другими словами, если исходное значение "истина", то "not" даст "ложь", а если исходное значение "ложь", то "not" даст "истину".


````python
a = not False
print(a)
````
