CSV (Comma Separated Values) - это формат хранения табличных данных, где каждая строка таблицы представляет собой отдельную запись, а значения каждой ячейки разделены запятой (или другим разделителем, например, точкой с запятой).

CSV файлы могут быть созданы и отредактированы в текстовых редакторах, электронных таблицах, базах данных и других приложениях. Они также могут быть легко импортированы во многие программы и использованы для анализа данных.


```python
import pandas as pd

# Open a CSV file using pandas library
df = pd.read_csv("movies.csv")

```

1. `import pandas as pd`

Здесь мы импортируем библиотеку pandas и называем ее pd (так, чтобы потом не писать каждый раз полное название).

2. `df = pd.read_csv("movies.csv")`

Создаем DataFrame (таблицу) под названием df и заполняем ее данными из CSV файла "movies.csv".

pd.read_csv() - функция, которая считывает данные из CSV файла и преобразует их в DataFrame.

"movies.csv" - название файла, из которого мы будем брать данные. Если файл находится не в той же директории, что и скрипт Python, нужно указать полный путь к файлу.

DataFrame - это основная структура данных в библиотеке Pandas. Он представляет собой таблицу с метками для строк и столбцов, похожую на таблицу в Excel или базу данных.

DataFrame может быть создан из различных источников данных, в том числе CSV, Excel, SQL или JSON файлов, массивов NumPy, списков словарей, HTML-таблиц и т.д.

# Определяем количество строк, столбцов в данных

1. `df_shape = df.shape`

Создаем переменную df_shape, которой присваиваем результат вызова функции shape() нашего DataFrame (таблицы) df.

pd.shape - функция, которая возвращает структуру с размерностью (количество строк, количество столбцов) DataFrame.

2. `num_rows = df_shape[0]`

Создаем переменную num_rows, которой присваиваем количество строк (равное первому элементу df_shape).

df_shape[0] - первый элемент (количество строк в DataFrame).

3. `num_columns = df_shape[1]`

Создаем переменную num_columns, которой присваиваем количество столбцов (равное второму элементу  df_shape).

df_shape[1] - второй элемент(количество столбцов в DataFrame).

4. `print(df_shape)`

Распечатываем размерность DataFrame

5. `print(num_rows)`

Распечатываем количество строк в DataFrame.

6. `print(num_columns)`

Распечатываем количество столбцов в DataFrame.

Таким образом, этот код позволяет получить размерность DataFrame и количество строк и столбцов в нем, что может использоваться для дальнейшей обработки и анализа данных.

#Находим и удаляем дубликаты строк

```python
# Drop duplicates
df = df.drop_duplicates(keep='first')
df
```

1. `df = df.drop_duplicates(keep='first')`

Этот код удаляет повторяющиеся строки (дубликаты) в DataFrame и сохраняет измененный DataFrame в переменной df.

pd.drop_duplicates(subset=None, keep='first', inplace=False, ignore_index=False) - это метод pandas, который удаляет дубликаты строк из DataFrame, то есть строки, в которых все значения для всех столбцов являются одинаковыми.

keep='first' - опция, которая указывает, какую строку оставить при удалении дубликатов. Если keep='first', то сохраняется первая строка дубликата, а остальные удаляются.

2. `df`

Печатаем на экран новый DataFrame df после удаления дубликатов.


#Извлекаем данные по определенному условию (например, все строки со значением в столбце больше определенного значения)

```python
df_threshold = df[df['Lead Studio'] == "Disney"]
print(df_threshold)
```

1. `df_threshold = df[df['Lead Studio'] == "Disney"]`

Выполняем фильтрацию данных из DataFrame и сохраняем результаты в новой переменной df_threshold.

2. `df['Lead Studio'] == "Disney"`

Создаем условие, которое проверяет, является ли значение в столбце "Lead Studio" равным "Disney".

3. `df[df['Lead Studio'] == "Disney"]`

Применяем это условие как фильтр к DataFrame df.

Результатом будет новый DataFrame df_threshold, содержащий только те строки, в которых значение "Lead Studio" равно "Disney".

Этот код может быть использован для извлечения данных только о фильмах, которые выпустила компания Disney.
