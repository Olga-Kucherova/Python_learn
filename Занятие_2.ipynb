# True/false
В Python Boolean (логические переменные) — это переменные, значение которых может быть либо «Истина», либо «Ложь».

Одно из наиболее распространенных применений логических значений в Python — это условные операторы, такие как `if` и `else`. *(этот персонаж повлияет на сюжет)*

````python
a = True
print(a)
````

Операторы сравнения.
Операторы сравнения в Python:
- "==" (равно) - проверяет, равны ли значения двух операндов;
- "!=" (не равно) - проверяет, не равны ли значения двух операндов;
- ">" (больше) - проверяет, является ли значение первого операнда больше, чем значение второго;
- ">=" (больше или равно) - проверяет, является ли значение первого операнда больше или равно значению второго;
- "<" (меньше) - проверяет, является ли значение первого операнда меньше, чем значение второго;
- "<=" (меньше или равно) - проверяет, является ли значение первого операнда меньше или равно значению второго.

Результатом любого оператора сравнения является булевское значение True или False, которое можно использовать в условных выражениях и в циклах. *(этот персонаж повлияет на сюжет)*

Например, выражение "5 > 2" вернет значение True, потому что 5 больше, чем 2. Выражение "3 == 3" также вернет True, потому что значения двух операндов равны друг другу.

````python
a = 6 > 5
print(a)
````

````python
a = 7 < 4
print(a)
````

````python
a = 7 == 7
print(a)
````

#If/else/elif

В Python if/else/elif являются условными конструкциями, которые выполняют определенный блок кода в зависимости от выполнения определенного условия.

if используется, чтобы определить условное выражение, а блок else выполняется в случае, если условие не выполняется.

Конструкция elif используется, чтобы добавить еще одно условие между if и else.

Пример использования конструкции if/else/elif в Python:
```
x = 10
if x > 5:
    print("x больше 5")
elif x == 5:
    print("x равно 5")
else:
    print("x меньше 5")
```

В этом примере первое условие (if x > 5) выполняется, поэтому на экран будет выведено сообщение "x больше 5". Если бы значение переменной x было равно 5, на экран было бы выведено сообщение "x равно 5". Если бы значение переменной x было меньше 5, на экран было бы выведено сообщение "x меньше 5".

Любое количество операторов elif можно добавлять к начальному оператору if, чтобы создать более сложную конструкцию. Обычно конструкцию if/else/elif в Python используют внутри функций или в циклах, чтобы выполнить различные действия в зависимости от значения переменной или результата выполнения какого-либо условия.

#and/or
В Python операторы логического "и" и "или" обычно обозначаются как "and" и "or". Они используются для проверки нескольких условий и возвращают булевское значение - True (истина) или False (ложь).

Оператор "и" (and) вернет True, если оба его операнда являются истинными. В противном случае, он вернет False.  (коньюнкция)

Оператор "или" (or) вернет True, если хотя бы один из его операндов является истинным. В противном случае, он вернет False. (дизъюнкция)

Таблица истинности операторов "и" и "или":

| Условие 1 | Оператор "и" | Условие 2 | Результат |
| --------- | ------------ | --------- | --------- |
| True      | and          | True      | True      |
| True      | and          | False     | False     |
| False     | and          | True      | False     |
| False     | and          | False     | False     |

| Условие 1 | Оператор "или" | Условие 2 | Результат |
| --------- | -------------- | --------- | --------- |
| True      | or             | True      | True      |
| True      | or             | False     | True      |
| False     | or             | True      | True      |
| False     | or             | False     | False     |

#Циклы. While
Цикл - это конструкция в программировании, которая позволяет выполнять некоторый блок кода несколько раз. Он исполняет команды до тех пор, пока выполняется некоторое условие.
Циклы позволяют выполнять повторяющиеся задачи или многократно обрабатывать большие объемы данных автоматически и эффективно, что делает программу более компактной. Существует два типа циклов в большинстве языков программирования: цикл for и цикл while.

````python
i = 0
while i < 5:
    print(i)
    i += 1
````

Цикл while используется в Python для выполнения блока кода до тех пор, пока условие истинно. Он продолжает итерирование, пока условие внутри while верно.
 Вот пример использования цикла while:

```
i = 0
while i < 5:
    print(i)
    i += 1
```

Этот код выводит числа от 0 до 4.
